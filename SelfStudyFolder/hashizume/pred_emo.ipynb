```python
import keras
from sklearn.svm import SVC
from gensim.models import word2vec 
import pickle
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# rep_doc
with open('./type_47/replaced_neg472_doc.pickle','rb') as f:
    re_n_doc = pickle.load(f)
with open('./type_47/replaced_pos472_doc.pickle','rb') as f:
    re_p_doc = pickle.load(f)
    
```

```python
w2v = word2vec.Word2Vec.load('./type_47/47type_300features_1minwords_10context_0ver')
```

```python
[x.split('-') for x  in w2v.wv.vocab if len(x.split('-'))>=2 ]
```

```python
re_w = word2vec.Word2Vec(re_n_doc+re_p_doc,sg=0,hs=1,size=300)
re_w2 = word2vec.Word2Vec(re_n_doc+re_p_doc,sg=1,hs=1,size=300)
```

```python
train_vocab = [x for x  in re_w.wv.vocab if len(x.split('-'))>=2 ]
```

```python
X = re_w[train_vocab]
Y = [1 if x.split('-')[-1] == 'p' else -1  for x in train_vocab]
x_tr, x_te, y_tr, y_te = train_test_split(X,Y)
```

```python
model = SVC(kernel='linear', random_state=None, max_iter=1)
model.fit(x_tr,y_tr)
pred_train = model.predict(x_te)
accuracy_train = accuracy_score(y_te, pred_train)
print('トレーニングデータに対する正解率： %.2f' % accuracy_train)
```

```python
score_list = []
X = re_w[train_vocab]
Y = [1 if x.split('-')[-1] == 'p' else -1  for x in train_vocab]
Y = np.array(Y)

from sklearn.model_selection import KFold
kfold = KFold(n_splits=10,shuffle=True,random_state=None)
#2222 0.64
for tr_index,te_index in kfold.split(X):
    x_train = X[tr_index]
    y_train = Y[tr_index]
    x_test = X[te_index]
    y_test = Y[te_index]

    model = SVC(kernel='rbf', random_state=1, max_iter=1,C=1000)
    model.fit(x_train,y_train)
    pred_train = model.predict(x_test)
    accuracy_train = accuracy_score(y_test, pred_train)
    print('トレーニングデータに対する正解率： %.2f' % accuracy_train)
    score_list.append(accuracy_train)
np.mean(score_list)
```
